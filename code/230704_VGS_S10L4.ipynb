```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import os
import pmdarima as pm 
from pmdarima.model_selection import train_test_split
import datetime
```

```python
os.chdir('P:\\repos\\03_edix\\S10_L4_entrega')
```

```python
df = pd.read_csv("data/spain_energy_market.csv", encoding='latin-1')
```

```python
df.sample(5)
```

```python
df = df[df['name']=="Demanda real"]
df = df.drop(['id', 'geoid', 'geoname', 'name'], axis=1)
df = df.reset_index(drop = True)
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df.dtypes
```

```python
plt.figure().set_figwidth(20)
plt.plot(df['datetime'], 
         df['value'])
plt.show()
```

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
```

```python
df.sample(5)
```

```python
df_month =  df.drop(['day'], axis=1) \
            .groupby(["year", 
                      "month"]) \
            ['value'].sum().reset_index()
df_month.sample(5)
```

```python
df_day =  df.groupby(["year", 
                      "month", 
                      "day"]) \
            ['value'].sum().reset_index()
df_day.sample(5)
```

```python
# autoarima model train
train, test = train_test_split(df_month['value'], train_size=int(0.8*df_month.shape[0]))
```

```python
# Fit a simple auto_arima model
arima = pm.auto_arima(train, 
                      error_action='ignore', 
                      trace=True,
                      suppress_warnings=True, 
                      maxiter=5,
                      seasonal=True, 
                      stationary=False, 
                      m=12) # monthly
```

```python
# Plot actual test vs. forecasts:
x = np.arange(test.shape[0])
plt.plot(x, test, color='red')
plt.plot(x, arima.predict(n_periods=test.shape[0]))
plt.title('Actual test samples vs. forecasts')
plt.legend(['Predicted', 'Actual'])
plt.show()
```
